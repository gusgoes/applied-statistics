```python
import numpy as np
import matplotlib.pyplot as plt

flips = np.random.binomial(1, 0.5, 1000)
flips
```

```python
#count the number of heads
heads = np.sum(flips)
heads
```

```python
#count the number of tails
tails = 1000 - heads
tails
```

```python
# plot the results
plt.bar(['heads', 'tails'], [heads, tails])
plt.show()
```

```python
#2 flips

two_flips = np.random.binomial(2, 0.5, 1000)
print(two_flips)
 #count the number of heads
heads = np.sum(two_flips)
print(heads)
#count the number of tails
tails = 2000 - heads
print(tails)
# plot the results
plt.bar(['heads', 'tails'], [heads, tails])
```

```python
#create histogram
fig, ax = plt.subplots()
ax.hist(two_flips, bins=3, edgecolor='black')
ax.set_title('Histogram of Two Flips')
ax.set_xlabel('Number of Heads')
ax.set_ylabel('Frequency')
ax.set_xticks([0, 1, 2])
plt.show()
```
